```python
#importar as bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# abrir o arquivo
dataset = pd.read_csv('resultados.csv')
dataset.head()
```

```python
# Comparar resultados dos modelos com Vgg
dataset_vgg = dataset[dataset['Extrator'] == 'Vgg']
sns.barplot(data=dataset_vgg, x=dataset['Modelo'], y='Precisao', hue='Tipo', legend='brief')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)
```

```python
# Comparar resultados do modelo com MobileNet
dataset_mobilenet = dataset[dataset['Extrator'] == 'MobileNet']
sns.barplot(data=dataset_mobilenet, x=dataset['Modelo'], y='Precisao', hue='Tipo', legend='brief')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)
```

```python
# comparar precisão entre os extratores
dataset.boxplot(column="Precisao", by="Extrator",grid=False)
```
